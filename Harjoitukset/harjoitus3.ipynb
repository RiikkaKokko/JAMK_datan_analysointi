```python
def emissions(url):
    # Kirjoita vastauksesi tähän
    # ...
    
    #lataa datan DataFrameen url:sta, jossa eroinmerkkinä on tabulaattori ('\t'),
    df = pd.read_csv(url, engine="python", encoding="utf-8-sig", sep="\\t")
    #print(df.head())
    
    #tallentaa DataFrameen sarakkeet 'country_code', 'main activity sector name', 'value' ja 'year',
    new_df = df[['main activity sector name', 'value', 'year"' ]].copy()
    #print(new_df.head())
    
    #poistaa DataFramesta rivit, jos sarakkeessa 'main activity sector name' esiintyy merkkijonot '20-99 All stationary installations' tai '21-99 All industrial installations (excl. combustion)'
    indexNames = new_df[(new_df['main activity sector name'] == '20-99 All stationary installations')].index 
    indexNames1 = new_df[(new_df['main activity sector name'] == '21-99 All industrial installations (excl. combustion)')].index 
    new_df.drop(indexNames , inplace=True)
    new_df.drop(indexNames1 , inplace=True)
    
    #print(new_df.head())
    
    #tallentaa uuteen DataFrameen kaikki rivit, joissa sarakkeen 'year' lukuarvo on yhtäsuuri tai suurempi kuin 2010 ja yhtäsuuri tai pienempi kuin 2015.
    new_df['year"'] = new_df['year"'].str.replace('"','', regex=True)
    new_df = new_df[~new_df['year"'].str.startswith('Total')]
    
    #print(new_df.head(50))
    new_df['year"'] = new_df['year"'].astype(int)
    
    new_df=(new_df[(new_df['year"']>= 2010) & (new_df['year"']<=2015)])
    #print(new_df.head())
    
    #laskee uuteen DataFrameen kuinka paljon päästöjä on ollut yhteensä vuosittain (laskee yhteen sarakkeen 'value' arvot, jotka on ryhmitelty sarakkeen 'year' arvojen mukaan),
    uusi_Dataframe=new_df.groupby('year"',as_index=False)['value'].sum()
    uusi_Dataframe.sort_values('year"')
    #print(uusi_Dataframe.head()) 
    
    #laskee uuteen DataFrameen uuteen sarakkeeseen 'change' kuinka paljon päästöt muuttuivat prosentuaalisesti edellisestä vuodesta (huom! vuoden 2010 arvo saa olla NaN, koska vuoden 2009 arvot on suodatettu pois),
    uusi_Dataframe['change'] = ((uusi_Dataframe['value'].diff())/uusi_Dataframe['value'].shift())*100
    #print(uusi_Dataframe.head()) 
    
    #pyöristää em. prosentuaalisen muutoksen yhden desimaalin tarkkuuteen,
    decimals = 1  
    uusi_Dataframe['change'] = uusi_Dataframe['change'].apply(lambda x: round(x, decimals))
    #print(uusi_Dataframe.head())
    
    #laskee DataFramen uuteen sarakkeeseen 'cumsum' kumulatiivisen päästöjen summan vuodesta 2010 vuoteen 2015,
    uusi_Dataframe['cumsum']=uusi_Dataframe['value'].cumsum()
    #print(uusi_Dataframe.head())
    
    #tulostaa edellä kuvatulla tavalla tuotetun DataFramen siten, että DataFramen indeksinä on 'year'.
    uusi_Dataframe.columns = ['year', 'emissions', 'change', 'cumsum']
    #print(uusi_Dataframe.head())
    
    uusi_Dataframe.set_index('year', inplace=True)
    #print(uusi_Dataframe.head())
    
    tulos=(uusi_Dataframe.head(6))
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv'
emissions(url)
```

```python
import pandas as pd
def emissions(url):
    # Kirjoita vastauksesi tähän
    # lataa datan DataFrameen url:sta, jossa eroinmerkkinä on tabulaattori ('\t'),
    df = pd.read_csv(url, engine="python", encoding="utf-8-sig", sep="\\t")
    #print(df.head())
    
    #tallentaa DataFrameen sarakkeet 'country_code', 'main activity sector name', 'value' ja 'year',
    new_df = df[['country_code', 'main activity sector name', 'value', 'year"' ]].copy()
    #print(new_df.head())
    
    #poistaa DataFramesta rivit, jos sarakkeessa 'main activity sector name' esiintyy merkkijonot '20-99 All stationary installations' tai '21-99 All industrial installations (excl. combustion)'
    indexNames = new_df[(new_df['main activity sector name'] == '20-99 All stationary installations')].index 
    indexNames1 = new_df[(new_df['main activity sector name'] == '21-99 All industrial installations (excl. combustion)')].index 
    new_df.drop(indexNames , inplace=True)
    new_df.drop(indexNames1 , inplace=True)
    #print(new_df.head())
    
    #tallentaa uuteen DataFrameen kaikki rivit, joissa sarakkeen 'year' lukuarvo on yhtäsuuri tai suurempi kuin 2010 ja yhtäsuuri tai pienempi kuin 2015.
    new_df['year"'] = new_df['year"'].str.replace('"','', regex=True)
    new_df = new_df[~new_df['year"'].str.startswith('Total')]
    
    #print(new_df.head(50))
    new_df['year"'] = new_df['year"'].astype(int)
    
    new_df=(new_df[(new_df['year"']>= 2010) & (new_df['year"']<=2015)])
    #print(new_df.head())
    
    #laskee uuteen DataFrameen kuinka paljon päästöjä kukin maa on tuottanut minäkin vuonna (laskee yhteen sarakkeen 'value' arvot, jotka on ryhmitelty sarakkeiden 'country_code' ja 'year' mukaan),
    uusi_Dataframe=new_df.groupby(["country_code", 'year"'], as_index=False)['value'].sum()
    uusi_Dataframe = pd.DataFrame(uusi_Dataframe)
    #print(uusi_Dataframe.head(53))
    
    #tallentaa uuteen DataFrameen kunkin maan (sarake 'country_code') toiseksi suurimmat päästöt (sarake 'value') ja päästön tuottamisvuoden (sarake 'year'),
    df1=uusi_Dataframe.groupby(['country_code'])[['value','year"']].apply(lambda x: x.nlargest(2, columns=['value'])) 
    #print(df1.head(5))
    
    dfuusi=df1.loc[df1.groupby('country_code')['value'].idxmin()]
    #print(dfuusi.head())
    
    #järjestää em. DataFramen rivit suurimmasta pienimpään sarakkeen 'value' mukaan,
    
    dfuusi = dfuusi.sort_values(by=['value'], ascending=False)
    #print(dfuusi.head())
    
    #nimeää sarakkeen 'country_code' sarakkeeksi 'ccode',
    dfuusi.reset_index(inplace=True)
    dfuusi.set_index('level_1', inplace=True)
    dfuusi.index.name = None
    #print(dfuusi.head())
    #print(list(dfuusi.columns))
    dfuusi.columns = ['ccode', 'value', 'year']
    #print(dfuusi.head())
    
    dfuusi = dfuusi[['ccode','year', 'value']]
    tulos=dfuusi.head()
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)

# Testaa funktiotasi:
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv'
emissions(url)
```
