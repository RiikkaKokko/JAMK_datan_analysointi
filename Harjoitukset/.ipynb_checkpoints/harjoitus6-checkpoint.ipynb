```python
import pandas as pd
import time
import datetime
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/saatiedot.csv'

def saa(url):
    # Kirjoita vastauksesi t√§h√§n alle:    

    #lataa datan DataFrameen url:sta,
    df=pd.read_csv(url)
    df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])
    df=df.sort_values(['Unnamed: 0'])
    df=df.reset_index(drop=True)#reset index or it really change order
    new_df = df.copy()

    #laskee DataFrameen uuden sarakkeen 'From start, s' riveille kuinka paljon aikaa on kulunut 
    #sekunteina aikaj√§rjestyksess√§ ensimm√§isest√§ mittaustapahtumasta,
    new_df['From start, s'] = new_df['Unnamed: 0'] - new_df['Unnamed: 0'][0]      

    #asettaa DataFramelle indeksin sarakkeesta 'From start, s' muodossa 'tuntia.minuuttia.sekuntia' Pandasin aikaleimana,
    def convert_timedelta(duration):
        days, seconds = duration.days, duration.seconds
        hours = days * 24 + seconds // 3600
        minutes = (seconds % 3600) // 60
        seconds = (seconds % 60)
        return datetime.datetime(1, 1, 1, hours,minutes,seconds).strftime('%H.%M.%S')

    new_df['hh.mm.ss_str']=new_df['From start, s'].apply(convert_timedelta)
    new_df['hh.mm.ss']=pd.to_datetime(new_df['hh.mm.ss_str'], format='%H.%M.%S')#.dt.time
    new_df = new_df.set_index(new_df['hh.mm.ss'])

    #tallentaa uuteen DataFrameen sarakkeet 'Air temperature, degC' ja 'Pressure (msl), hPa',
    df1 = new_df[['hh.mm.ss', 'Air temperature, degC', 'Pressure (msl), hPa']].copy()
    #laskee edelleen uuteen DataFrameen sarakkeiden 'Air temperature, degC' ja 'Pressure (msl),
    #hPa' riveill√§ olevien arvojen tuntikohtaiset keskiarvot,

    #df_mean=df1.groupby(pd.Grouper(freq='H'))['Air temperature, degC', 'Pressure (msl), hPa'].mean()
    df_mean=df1.groupby(pd.Grouper(freq='H'))[['Air temperature, degC', 'Pressure (msl), hPa']].mean()
    df_mean=df_mean[df_mean.index>=pd.to_datetime('07.00.00', format='%H.%M.%S')]
    temp_ind=df_mean.index.strftime('%H.%M.%S')
    df_mean=df_mean.set_index(temp_ind)
    df_mean=df_mean.applymap("{:.1f}".format)
    tulos=(df_mean.head(5))
    
    print(tulos)

saa(url)
```

```python
import pandas as pd
import numpy as np
import datetime as dt
import math
import datetime

start='2019-12-31 23:59:00'
end='2020-01-01 00:00:30'
def timedata(start, end):
    # Kirjoita vastauksesi t√§h√§n alle:
    
    #muodostaa DataFramen indeksiksi Pandasin aikaleiman muodossa tuntia:minuuttia:sekuntia 20 sekunnin v√§lein alkaen ajanhetkest√§ '2019-12-31 23:59:00' hetkeen '2020-01-01 00:00:30',
    rng = pd.date_range(start, end, freq='20s')
    df = pd.DataFrame({ 'Date': rng }) 
    #df['Date'] = pd.to_datetime(df['Date'])
    df['new_date'] = [d.date() for d in df['Date']]
    df['new_time'] = [d.time() for d in df['Date']]
    #print(df.head())
    #print(df.shape)
    
    #esitt√§√§ uudessa DataFramessa kappaleen lentoajan  ùë°  sekunteina sarakkeessa 't, s',
    new_df = df.copy()
    #print(new_df.head())
    #new_df['t, s'] = pd.to_timedelta(new_df['Date'] - new_df['Date'][0])
    new_df['t, s'] = pd.to_timedelta(new_df['Date'] - new_df['Date'][0]).astype('timedelta64[s]')
    #print(new_df.head())
    
    #laskee DataFrameen kappaleen paikan x-akselin suunnassa metrein√§ sarakkeeseen 'x, m'
    new_df['x, m']=(new_df['t, s']*600)/(math.sqrt(2))
    #print(new_df.head())
    
    #laskee DataFrameen kappaleen paikan y-akselin suunnassa metrein√§ sarakkeeseen 'y, m',
    new_df['y, m']=new_df['x, m']-(9.82*(new_df['t, s']**2)/2)
    #print(new_df.head())
    
    #py√∂rist√§√§ ja esitt√§√§ DataFramessa lukuarvot kokonaislukujen tarkkuuteen,
   
    #new_df['t, s']=new_df['t, s'].astype(int)
    #new_df['x, m']=new_df['x, m'].astype(int)
    #new_df['y, m']=new_df['y, m'].astype(int)
    #print(new_df.head())
    
    new_df=new_df.drop(['Date', 'new_date'], axis = 1)
    new_df = new_df.set_index(new_df['new_time'])
    new_df=new_df.drop(['new_time'], axis = 1)
    new_df.index.names = ['hh:mm:ss']
    new_df=new_df.applymap("{:.0f}".format)
    tulos=new_df.head(5)
    print(tulos)
    
    
    
    # √Ñl√§ kirjoita t√§h√§n muita print komentoja ennen viimeist√§ print(tulos) komentoa, koska t√§ll√∂in vastauksesi saatetaan tulkita v√§√§rin.
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    #print(tulos) 
    
timedata(start, end)
```

## 4. Aikasarjadatan lineaarinen interpolointi

T√§m√§n teht√§v√§n datassa esiintyv√§t sarakkeet 'hh:mm:ss', 'aika, s' ja 'y, m', mitk√§ edustavat kellonaikaa mittauksian alusta, aikaa mittauksien alusta sekunteina ja kappaleen paikkaa metrein√§. Tee funktio interp_times(url), joka tekee seuraavaa:
* lataa datan [url:sta](https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/motion.csv) DataFrameen,
* t√§ytt√§√§ mittauksien alusta l√§htien uuden DataFramen indekseihin puuttuvat kellonajan arvot ja puuttuvat ajan arvot sarakkeeseen 't_tot, s' 5 sekunnin aikav√§lein,
* t√§ytt√§√§ mittauksien alusta l√§htien uuteen DataFrameen puuttuvat kappaleen paikan arvot sarakkeeseen 'int_y, m' lineaarisen approksimaation mukaan 5 sekunnin v√§lein,
* py√∂rist√§√§ DataFramen lukuarvot kokonaislukujen tarkkuuteen,
* tulostaa vastaukseen sarakkeet 't_tot, s' ja 'int_y, m', kun mittauksien alusta lukien kellonaika on '00:00:45'-'00:01:31'.

T√§ss√§ teht√§v√§ss√§ lineaarinen approksimaatio tarkoittaa kahden eri pisteen v√§liin j√§√§vien paikan $y$ arvojen laskemista suoran yht√§l√∂n avulla
$$
y=\left(\frac{y_{i+1}-y_{i}}{t_{i+1}-t_{i}}\right)\cdot(t-t_{i})+y_{i},
$$
miss√§ $y_i$ ja $y_{i+1}$ ovat alkuper√§isen datasetin per√§kk√§iset kappaleen paikat (metrein√§) sek√§ $t_i$ ja $t_{i+1}$ ovat alkuper√§isen datasetin per√§kk√§iset ajanhetket sekunteina. Kappaleen paikka $y$ on metrein√§ ja tunnettujen pisteiden v√§liss√§ oleva ajanhetki $t$ on sekunteina. Indeksi $i$ tarkoittaa alkuper√§isen datan rivi-indeksi√§ ja $i+1$ tarkoittaa seuraavaa rivi-indeksi√§.

```python
import pandas as pd
import numpy as np
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/motion.csv'

def interp_times(url):
    # Kirjoita vastauksesi t√§h√§n alle:
    
    #lataa datan url:sta DataFrameen,   
    #df=pd.read_csv(url)
    df =pd.read_csv(url, sep = "\t")
    df['hh:mm:ss'] = pd.to_datetime(df['hh:mm:ss'])

  
    #print(df.head(20))
    
    #t√§ytt√§√§ mittauksien alusta l√§htien uuden DataFramen indekseihin puuttuvat kellonajan arvot ja puuttuvat ajan arvot sarakkeeseen 't_tot, s' 5 sekunnin aikav√§lein,
    df1 = df.set_index('hh:mm:ss').resample('5s').mean()
    #print(df1.head())
    
    #t√§ytt√§√§ mittauksien alusta l√§htien uuteen DataFrameen puuttuvat kappaleen paikan arvot sarakkeeseen 'int_y, m' lineaarisen approksimaation mukaan 5 sekunnin v√§lein,
    #df1=df1.fillna(method='ffill')
    df1=df1.interpolate()
 
   #print(df1.head(20))
    
    #py√∂rist√§√§ DataFramen lukuarvot kokonaislukujen tarkkuuteen,
    df1=df1.applymap("{:.0f}".format)
    #print(df1.head(20))
    
    #tulostaa vastaukseen sarakkeet 't_tot, s' ja 'int_y, m', kun mittauksien alusta lukien kellonaika on '00:00:45'-'00:01:31'.
    temp_ind=df1.index.strftime('%H:%M:%S')
    df1=df1.set_index(temp_ind)
    #df1=((df1[df1.index>'00.00.44']))
    tulos=df1.loc[(df1.index>'00:00:44') & (df1.index<'00:01:31')]
    tulos.columns = ['t_tot, s', 'int_y, m']
    

    
    
    # √Ñl√§ kirjoita t√§h√§n muita print komentoja ennen viimeist√§ print(tulos) komentoa, koska t√§ll√∂in vastauksesi saatetaan tulkita v√§√§rin.
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    print(tulos)
    
interp_times(url)
```
