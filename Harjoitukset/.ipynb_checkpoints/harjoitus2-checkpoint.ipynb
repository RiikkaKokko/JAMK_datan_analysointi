```python
import pandas as pd
import numpy as np
# Palautuksesi alkaa tästä
def suomisahko(url):
    
    # Kirjoita vastauksesi tähän:
    
    #Lukee url:stä datan DataFrameen kun sarakkeiden erottimena on pilkku "," ja desimaalierottimena piste ",",
    df = pd.read_csv(url, sep=",", decimal=",")
    #print(df.head())
    
    
    #jos sarakkeissa 'Year Total €/MWh' ja 'December €/MWh' niiden välillä on merkki ".", korvataan merkki numpyn nan arvolla hyödyntäen pandasin isin() metodia,
    columns=df.iloc[:, 2:15]
    for col in columns:
        df[col]=np.where(df[col].isin(['.']), np.nan, df[col])
    #print(df.head(5))
        
    
    #muuntaa sarakkeiden 'Year Total €/MWh' ja 'December €/MWh' ja niiden välissä olevien sarakkaiden tietotyypiksi float,
    
    for i in df.columns[2:15]:
        #print(i)
        df[i]=df[i].astype(float)

    #print(df.dtypes)
    
    #kopioi DataFramesta kaikki sarakkeet ja ne rivit noiden sarakkeessa 'Price area' on merkkijono 'Finland',
    #print(df.columns)

    #Finland=(df.loc[df['Price area'] == 'Finland'])
    Finland=(df.loc[df['Price area'] == 'Finland']).copy()
    #print(Finland)
    
    #laskee koko vuoden sähkön hinnan keskiarvon uuteen sarakkeeseen 'Average €/MWh' kuukausilta ('January €/MWh'-'December €/MWh'),
    Januarydecember=Finland.iloc[:, 3:15]
    
    #for i in Finland.columns[3:15]:
     #   print(i)
      #  Finland['Average €/MWh'] = Finland[[i]].mean(axis=1)
    
    #print(January_December)
    Finland['Average €/MWh'] = Januarydecember.mean(axis=1)
    #print(Finland)
    
    #laskee sarakkeeseen 'Below average price' kuinka monena kuukautena sähkönhinta on ollut alle vuoden keskiarvon
    Finland["Below average price"]=Finland.apply(lambda x: (x[3:15]<x[15]).sum(), axis=1)
    
    #laskee sarakkeeseen 'Above average price' kuinka monena kuukautena sähkönhinta on ollut yli vuoden keskiarvon tai yhtä paljon kuin keskiarvohinta,
    Finland["Above average price"]=Finland.apply(lambda x: (x[3:15]>x[15]).sum(), axis=1)
   
    
    #tulostaa vain ne kymmenen riviä (head(10)), jolloin sarakkeiden 'Below average price' ja 'Above average price' summa on 12.
    Finland["sums"]=Finland.apply(lambda x: (x[16] + x[17]), axis=1)
    
    new_df = Finland[['Year', 'Below average price', 'Above average price', 'sums']].copy()
    
    rslt_df = new_df[new_df['sums'] == 12]  
    #print(rslt_df)
    
    tulos1=rslt_df.drop('sums', axis=1)
    #print(tulos1)
    
    tulos=tulos1.head(10)
    
    # Vastauksesi tulostetaan
    print(tulos)

# Palautuksesi päättyy tähän.
    
url = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta2/harjoitukset/ehi_010_2020q2.csv"
# Voit kokeilla funktiotasi seuraavasti:
suomisahko(url)
```

```python
import pandas as pd
def mail(url):
    # Kirjoita vastauksesi tähän alle:
    
    # lataa url:stä DataFrameen henkilötietodatan, jossa erotinmerkkinä on tabulaattori ('\t'),

    df = pd.read_csv(url, engine="python", encoding="utf-8-sig", sep="\\t")
    #print(df.tail(50))
    
    #tallentaa uuteen DataFrameen sarakkeesta 'NameOfContactPerson' sukunimen ja sarakkeesta 'Email' sähköpostiosoitteen,
    #df['len']=df.NameOfContactPerson.apply(lambda x: len(x.split()))
    
    def get_name(name):
        last_name=''
        splits=name.split(" ", 3)
        if len(splits)==1:
            last_name=splits[0]
        elif len(splits)==2:
            last_name=splits[1]
        elif len(splits)==3:
            last_name=splits[2]
        elif len(splits)==4:
            last_name=splits[3]
        return last_name
    
    df["NameOfContactPerson"]=df.NameOfContactPerson.apply(get_name)
    #print(df.head(50))
    
    new_df = df[["NameOfContactPerson",'EMail"']].copy()
    #print(new_df.head(50))
    new_df['EMail"'] = new_df['EMail"'].str.replace('"','', regex=True)
    
    
    #poistaa rivit, joissa 'NameOfContactPerson' sarakkeessa on samat arvot siten, että järjestyksessä ensimmäinen rivi jätetään,
    #new_df=new_df.groupby((new_df["NameOfContactPerson"] != new_df["NameOfContactPerson"].shift()).cumsum().values).first()
    new_df = new_df.drop_duplicates(subset=["NameOfContactPerson"], keep='first')
    
    #järjestää yhteystiedot sukunimen mukaan aakkosissa suurimmasta pienimpään(Ö->A),
    new_df=new_df.sort_values('NameOfContactPerson', ascending=False)
    new_df.columns = ['sukunimi', 'email']
    tulos=new_df.head(5)
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    print(tulos)
# Palautuksesi päättyy tähän.    

# Testaa funktiotasi:
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta2/harjoitukset/osotteita.csv'
mail(url)
```

```python
import pandas as pd
import numpy as np
def mail(url):
    # Kirjoita vastauksesi tähän alle:
    
    #lataa url:stä DataFrameen henkilötietodatan, jossa erotinmerkkinä on tabulaattori ('\t'),
    df = pd.read_csv(url, engine="python", encoding="utf-8-sig", sep="\\t")
    #print(df.head())
    
    #poistaa rivit, joissa 'NameOfContactPerson' sarakkeessa on samat arvot siten, että järjestyksessä ensimmäinen rivi jätetään,
    df = df.drop_duplicates(subset=["NameOfContactPerson"], keep='first')
    
    #print(df.head(55))
    #poistaa sarakkeen 'NameOfContactPerson' kaikilta riveiltä kaarisulkeet ja niiden sisällä olevat merkkijonot,
    df['NameOfContactPerson'] = df['NameOfContactPerson'].str.replace(r"(\s*\(.*?\)\s*)", " ").str.strip()
    
    #print(df.head(50))
    s
    #tallentaa uuteen DataFrameen sarakkeesta 'NameOfContactPerson' sukunimen ja sarakkeesta 'Email' sähköpostiosoitteen,
    def get_name(name):
        last_name=''
        splits=name.split(" ", 3)
        if len(splits)==1:
            last_name=splits[0]
        elif len(splits)==2:
            last_name=splits[1]
        elif len(splits)==3:
            last_name=splits[2]
        elif len(splits)==4:
            last_name=splits[3]
        return last_name
    
    df["NameOfContactPerson"]=df.NameOfContactPerson.apply(get_name)
    #print(df.tail(60))
    
    #print(df.head(20))
    new_df = df[["NameOfContactPerson",'EMail"']].copy()
    #print(new_df.head(50))
    
    
    #tallentaa uuteen DataFramen sarakkeeseen ('surname') vain ne rivit, joissa sukunimi sarakkeessa esiintyy aakkosia tai yhdysviiva "-" osana sukunimeä,
    new_df=new_df[~new_df['NameOfContactPerson'].str.contains(r'[0-9]')]
    #print(new_df.tail(60))
    
    new_df.columns = ['surname', 'EMail"']
    #print(new_df.tail(60))
    
    #tallentaa uuteen DataFrameen sarakkeen ('domainend'), jossa on sähköpostiosoitteen domainin loppuosa (esim. fi),
    #print(df.head(50))
    new_df['EMail"'] = new_df['EMail"'].str.replace('"','', regex=True)
    
    new_df['domainend'] =new_df['EMail"'].apply(lambda x: x.split('.')[-1])
    
    
    #poistaa sarakkeesta 'domainend' tyhjät merkkijonot,
    nan_value = float("NaN")
    new_df.replace("", nan_value, inplace=True)
    new_df.dropna(subset = ["domainend"], inplace=True)
    #print(new_df.head(50))
    
    #järjestää DataFramen rivit sarakkeen 'surname' mukaan aakkosittain nousevaan järjestykseen (A -> Ö)
    new_df=new_df.sort_values('surname', ascending=True)
    
    #print(new_df.head(50))
    new_df = new_df.drop('EMail"', axis=1)
    tulos=(new_df.head(5))
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    
    print(tulos)

# Palautuksesi päättyy tähän.

# Testaa funktiotasi:
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta2/harjoitukset/osotteita.csv'
mail(url)
```
